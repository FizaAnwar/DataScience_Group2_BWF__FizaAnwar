Set a specific column as the index of the DataFrame.


```python
df.set_index(	'PassengerId', inplace=True)

# Display the DataFrame
print(df)
```

Select a specific column and display its values.


```python
specific_column = df['Survived']

# Display the Survived
print(specific_column)
```

Select multiple columns and display the resulting DataFrame.


```python
selected_columns = df[['Name', 'Sex','Survived','Pclass']]

# Display the resulting DataFrame
print(selected_columns)
```

Select a subset of rows using the .loc method.


```python
subset = df.loc[(df['Sex'] == 'female') & (df['Survived'] == 1)]

# Display the result of DataFrame
print(subset)
```

Select a subset of rows and columns using the .iloc method.


```python
subset = df.iloc[:10, :3]
print(subset)
```

Filter rows based on a condition.


```python
older_first_class_passengers = df[(df['Age'] > 55) & (df['Pclass'] == 1)]
print(older_first_class_passengers)
```

Group the DataFrame by a specific column and calculate the mean of each group.


```python
mean_age_fare_by_pclass = df.groupby('Pclass')[['Age', 'Fare']].mean()
print(mean_age_fare_by_pclass)
```

Group the DataFrame by multiple columns and calculate the sum of each group.


```python
sum_survived_fare_by_pclass_sex = df.groupby(['Pclass', 'Sex'])[['Survived', 'Fare']].sum()
print(sum_survived_fare_by_pclass_sex)
```

Use the agg method to apply multiple aggregation functions to grouped data.


```python
def my_custom_function(values):
    # Custom aggregation function example
    return values.max() - values.min()

# Example: Apply a custom aggregation function
agg_functions_custom = {
    'Fare': my_custom_function
}

result_custom = df.groupby(['Pclass', 'Sex']).agg(agg_functions_custom)
print(result_custom)
```

Calculate the size of each group.

```python
group_size = df.groupby(['Pclass', 'Sex']).size()
print(group_size)
```

Select rows based on multiple conditions.


```python
filtered_df = df[(df['Sex'] == 'female') & (df['Age'] > 30) & (df['Survived'] == 1)&(df['Pclass']==1)]
print(filtered_df.head())
```

Use the query method to filter rows.

```python
filtered_df_query = df.query("Sex == 'female' and Age > 30")
print(filtered_df_query.head())
```

Use isin to filter rows based on a list of values.

```python
pclass_values = [1, 2]
filtered_df_isin = df[df['Pclass'].isin(pclass_values)]
print(filtered_df_isin.head())
```

Select specific columns and rename them:

```python
change_columns = ['Name', 'Sex', 'Age']
renamed_df = df[change_columns].rename(columns={'Name': 'PassengerName', 'Sex': 'Gender'})
print(renamed_df.head())
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df1=pd.read_csv("/content/Cleaned Car.csv")
df1.head()
```

```python
# Setting the style
sns.set_style('whitegrid')
# Scatter plot
sns.scatterplot(data=df1, x='Price', y='year', palette='viridis')
plt.title('Scatter Plot of Price vs Year')
plt.xlabel('Price')
plt.ylabel('year')
plt.show()
```

```python
# Setting the style
sns.set_style('darkgrid')

# Line plot
sns.lineplot(data=df1, x='company', y='Price', palette='cubehelix')
plt.title('Line Plot of company vs Price')
plt.xlabel('Company')
plt.ylabel('Price')
plt.show()
```

```python
# Setting the style
sns.set_style('white')

# Bar plot
sns.barplot(data=df1, x='Price', y='fuel_type',palette='muted' )
plt.title('Bar Plot of Price of Car by Fuel Type')
plt.xlabel('Price')
plt.ylabel('fuel_type')
plt.show()
```

```python
# Setting the style
sns.set_style('ticks')

# Count plot
sns.countplot(data=df1, x='kms_driven',palette='pastel')
plt.title('Count Plot of Kms_Driven by Car')
plt.xlabel('Kms_Driven')
plt.ylabel('Count')
plt.show()
```

```python
# Setting the style
sns.set_style('dark')

# Box plot
sns.boxplot(data=df1, x='year', y='Price', palette='deep')
plt.title('Box Plot of Total Price of car by year')
plt.xlabel('year')
plt.ylabel('Price')
plt.show()
```

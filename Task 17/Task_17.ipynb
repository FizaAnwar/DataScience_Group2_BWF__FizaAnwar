Subplots:


Generate a figure with 2x2 subplots, each containing a different type of plot (e.g., line plot, scatter plot, bar plot, and histogram).


```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
x = np.linspace(0, 10, 100)
y = np.sin(x)
categories = ['A', 'B', 'C', 'D']
values = [10, 15, 7, 10]
data = np.random.randn(1000)

# Create a 2x2 subplot
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Line plot
axs[0, 0].plot(x, y, color='blue')
axs[0, 0].set_title('Line Plot')

# Scatter plot
axs[0, 1].scatter(x, y, color='red')
axs[0, 1].set_title('Scatter Plot')

# Bar plot
axs[1, 0].bar(categories, values, color='green')
axs[1, 0].set_title('Bar Plot')

# Histogram
axs[1, 1].hist(data, bins=30, color='purple')
axs[1, 1].set_title('Histogram')
plt.tight_layout()
plt.show()
```

Customize the subplots with titles, axis labels, and legends.


```python
# Sample data
x = np.linspace(0, 10, 100)
y = np.sin(x)
categories = ['A', 'B', 'C', 'D']
values = [10, 15, 7, 10]
data = np.random.randn(1000)

# 2x2 subplot
fig, axs = plt.subplots(2, 2, figsize=(12, 10))

# Line plot
axs[0, 0].plot(x, y, color='blue', label='sin(x)')
axs[0, 0].set_title('Line Plot')
axs[0, 0].set_xlabel('X-axis')
axs[0, 0].set_ylabel('Y-axis')
axs[0, 0].legend()

# Scatter plot
axs[0, 1].scatter(x, y, color='red', label='sin(x) points')
axs[0, 1].set_title('Scatter Plot')
axs[0, 1].set_xlabel('X-axis')
axs[0, 1].set_ylabel('Y-axis')
axs[0, 1].legend()

# Bar plot
axs[1, 0].bar(categories, values, color='green', label='Values')
axs[1, 0].set_title('Bar Plot')
axs[1, 0].set_xlabel('Categories')
axs[1, 0].set_ylabel('Values')
axs[1, 0].legend()

# Histogram
axs[1, 1].hist(data, bins=30, color='purple', label='Data')
axs[1, 1].set_title('Histogram')
axs[1, 1].set_xlabel('Value')
axs[1, 1].set_ylabel('Frequency')
axs[1, 1].legend()
plt.tight_layout()
plt.show()
```

Histograms:
Create a histogram for a given numerical data set. Customize the number of bins.


```python
# Sample data
data = np.random.randn(100)

# Number of bins
num_bins = 30

# Create histogram
plt.figure(figsize=(6, 6))
plt.hist(data, bins=num_bins, color='blue', edgecolor='black', alpha=0.7)

# Customizing
plt.title('Histogram of Data')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

Plot multiple histograms on the same figure to compare different data sets.

```python
# Sample data
data1 = np.random.randn(1000)
data2 = np.random.randn(1000) + 1
data3 = np.random.randn(1000) - 2

# Number of bins
num_bins = 20

# Create figure and axis
plt.figure(figsize=(8, 6))

# Plot histograms
plt.hist(data1, bins=num_bins, alpha=0.5, label='Data-Set 1', color='blue', edgecolor='black')
plt.hist(data2, bins=num_bins, alpha=0.5, label='Data-Set 2', color='red', edgecolor='black')
plt.hist(data3, bins=num_bins, alpha=0.5, label='Data-Set 3', color='cyan', edgecolor='black')

# Customizing
plt.title('Comparison of Multiple Data Sets')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend(loc='upper right')
plt.grid(True)
plt.show()
```

Normalize the histograms and add appropriate labels and titles.


```python
# Sample data
data1 = np.random.randn(1000)
data2 = np.random.randn(1000) + 3
data3 = np.random.randn(1000) - 5

# Number of bins
num_bins = 20

# Create figure
plt.figure(figsize=(10, 6))

# Plot normalized histograms
plt.hist(data1, bins=num_bins, alpha=0.3, label='Data Set 1', color='blue', edgecolor='black', density=True)
plt.hist(data2, bins=num_bins, alpha=0.4, label='Data Set 2', color='red', edgecolor='black', density=True)
plt.hist(data3, bins=num_bins, alpha=0.5, label='Data Set 3', color='green', edgecolor='black', density=True)

# Customizing
plt.title('Normalized Comparison of Multiple Data Sets')
plt.xlabel('Value')
plt.ylabel('Density')
plt.legend(loc='upper right')
plt.grid(True)
plt.show()
```

Box Plots:

Create a box plot for a given numerical data set.

```python
data = np.random.randn(100)
plt.figure(figsize=(6, 6))
plt.boxplot(data)

# Customizing the plot
plt.title('Box Plot of Data')
plt.xlabel('Data Set')
plt.ylabel('Value')
plt.grid(True)

plt.show()
```

Generate multiple box plots in a single figure to compare different data sets.


```python
# Sample data
data1 = np.random.randn(100)
data2 = np.random.randn(100) + 5
data3 = np.random.randn(100) - 3
data = [data1, data2, data3]


plt.figure(figsize=(5, 6))
plt.boxplot(data, patch_artist=True)

# Customizing the plot
plt.title('Comparison of Multiple Data Sets')
plt.xlabel('Data Set')
plt.ylabel('Value')
plt.xticks([2, 3, 4], ['Data Set 1', 'Data Set 2', 'Data Set 3'])
plt.grid(True)

plt.show()
```

Customize the box plots with titles, labels, and change the colors.


```python
# Sample data
data1 = np.random.randn(100)
data2 = np.random.randn(100) + 4
data3 = np.random.randn(100) - 3

data = [data1, data2, data3]
plt.figure(figsize=(4, 6))

boxprops = dict(facecolor='lightblue', color='blue')
medianprops = dict(color='red', linewidth=2)
whiskerprops = dict(color='black', linewidth=1.5)
capprops = dict(color='black', linewidth=1.5)
flierprops = dict(marker='o', markerfacecolor='green', markersize=5, linestyle='none')

plt.boxplot(data, patch_artist=True, boxprops=boxprops, medianprops=medianprops,
            whiskerprops=whiskerprops, capprops=capprops, flierprops=flierprops)

# Customizing
plt.title('Comparison of Multiple Data Sets')
plt.xlabel('Data Set')
plt.ylabel('Value')
plt.xticks([2, 4, 6], ['Data Set 1', 'Data Set 2', 'Data Set 3'])
plt.grid(True)


plt.show()
```

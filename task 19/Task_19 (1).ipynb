```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Example dataset
df = sns.load_dataset("penguins")
df.head()


```

```python
df['bill_length_mm'].fillna(df['bill_length_mm'].mean(), inplace=True)
df['bill_depth_mm'].fillna(df['bill_depth_mm'].mean(), inplace=True)
df['flipper_length_mm'].fillna(df['flipper_length_mm'].mean(), inplace=True)
df['body_mass_g'].fillna(df['body_mass_g'].mean(), inplace=True)
categorical_columns = ['sex']
for column in categorical_columns:
    df[column].fillna('Male', inplace=True)
df.head()
```

```python
# Setting the style
sns.set_style('whitegrid')

# Linear Regression Plot
sns.regplot(data=df, x='bill_length_mm', y='bill_depth_mm')
plt.title('Linear Regression Plot of bill_length_mm vs bill_depth_mm')
plt.xlabel('bill_length_mm')
plt.ylabel('bill_depth_mm')
plt.show()
```

```python
# Multiple Linear Regression Plot
sns.lmplot(data=df, x='bill_length_mm', y='flipper_length_mm', hue='sex')
plt.title('Multiple Linear Regression Plot of bill_length_mm vs flipper_length_mm by Sex')
plt.xlabel('Bill_Length_mm')
plt.ylabel('Flipper_Length_mm')
plt.show()
```

```python
# Residual Plot
sns.residplot(data=df, x='bill_length_mm', y='body_mass_g')
plt.title('Residual Plot of Bill_length_mm vs body_mass_g')
plt.xlabel('Bill_length_mm')
plt.ylabel('Body_mass_g')
plt.show()
```

```python
# Histogram
sns.histplot(data=df, x='species', bins=30, kde=True)
plt.title('Histogram of species')
plt.xlabel('Species')
plt.ylabel('Frequency')
plt.show()
```

```python
# Histogram
sns.histplot(data=df, x='island', bins=30, kde=True)
plt.title('Histogram of island')
plt.xlabel('island')
plt.ylabel('Frequency')
plt.show()
```

```python
# KDE Plot
sns.kdeplot(data=df, x='body_mass_g', shade=True)
plt.title('KDE Plot of Body_mass_g')
plt.xlabel('body_mass_g	')
plt.ylabel('Density')
plt.show()
```

```python
# Distribution Plot
sns.histplot(data=df, x='bill_depth_mm', kde=True)
plt.title('Distribution Plot of Bill_depth_mm	')
plt.xlabel('Bill_depth_mm	')
plt.ylabel('Density')
plt.show()
```

```python
# Box Plot
sns.boxplot(data=df, x='species', y='bill_length_mm')
plt.title('Box Plot of bill_length_mm	by Species')
plt.xlabel('Species')
plt.ylabel('Bill_length_mm')
plt.show()
```

```python
# Violin Plot
sns.violinplot(data=df, x='sex', y='bill_depth_mm', hue='sex', split=True)
plt.title('Violin Plot of Bill_depth_mm by Sex')
plt.xlabel('Sex')
plt.ylabel('Bill_depth_mm')
plt.show()
```

```python
# Pair Plot
sns.pairplot(df, hue='sex')
plt.suptitle('Pair Plot of Species Dataset', y=1.02)
plt.show()
```

```python

for column in df.columns:
    # Attempt to convert the column to numeric, coerce errors to NaN
    df['species'] = pd.to_numeric(df['species'], errors='coerce')

df.info()
```

```python
# Select only numeric columns
numeric_df = df.select_dtypes(include=[float, int])

# Compute the correlation matrix
corr = numeric_df.corr()


# Heatmap
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Heatmap of species dataset Correlation Matrix')
plt.show()

```

```python
# Joint Plot
sns.jointplot(data=df, x='bill_length_mm', y='island', kind='reg')
plt.suptitle('Joint Plot of bill_length_mm vs island', y=1.02)
plt.show()
```

```python
# Facet Grid
g = sns.FacetGrid(df, col='sex', row='island')
g.map(sns.scatterplot, 'bill_length_mm', 'flipper_length_mm')
g.add_legend()
plt.show()
```

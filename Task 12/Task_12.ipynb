Importing Libraries

```python
import pandas as pd
import numpy as np
```

Create a Pandas Series from a Python list, numpy array, and a dictionary.

```python
python_list = [20,67,4,5,6,8]


series_from_list = pd.Series(python_list)
print("Series from list:")
print(series_from_list)
```

```python

numpy_array = np.array([100, 75, 30, 40, 50])


series_from_array = pd.Series(numpy_array)
print("\nSeries from numpy array:")
print(series_from_array)
```

```python
dictionary = {'a': 100, 'b': 200, 'c': 300, 'd': 400, 'e': 500}

series_from_dict = pd.Series(dictionary)
print("\nSeries from dictionary:")
print(series_from_dict)
```

Assign a custom index to the Series.

```python
# Python list
python_list = [1, 2, 3, 4, 5]
custom_index_list = ['a', 'b', 'c', 'd', 'e']

# Numpy array
numpy_array = np.array([10, 20, 30, 40, 50])
custom_index_array = ['A', 'B', 'C', 'D', 'E']

# Dictionary
dictionary = {'a': 100, 'b': 200, 'c': 300, 'd': 400, 'e': 500}
custom_index_dict = ['a', 'b', 'c', 'd', 'e']

series_from_list = pd.Series(python_list, index=custom_index_list)
series_from_array = pd.Series(numpy_array, index=custom_index_array)
series_from_dict = pd.Series(dictionary).reindex(custom_index_dict)


print("Series from list with custom index:")
print(series_from_list)
print("\nSeries from numpy array with custom index:")
print(series_from_array)
print("\nSeries from dictionary with custom index:")
print(series_from_dict)
```

Basic operations on series

```python
series1 = pd.Series([1, 2, 3, 4, 5], index=['a', 'b', 'c', 'd', 'e'])
series2 = pd.Series([50,60,70,80,90], index=['a', 'b', 'c', 'd', 'e'])

print("Series 1:")
print(series1)
print("\nSeries 2:")
print(series2)

# Addition
addition = series1 + series2
print("\nAddition:")
print(addition)

# Subtraction
subtraction = series1 - series2
print("\nSubtraction:")
print(subtraction)

# Multiplication
multiplication = series1 * series2
print("\nMultiplication:")
print(multiplication)

# Division
division = series1 / series2
print("\nDivision:")
print(division)
```

Access elements using index labels and positions.

```python
# Sample Series
series = pd.Series([10, 20, 30, 40, 50], index=['a', 'b', 'c', 'd', 'e'])

print("Series:")
print(series)

# Accessing elements using index labels:
element_a = series['a']
print("\nElement at index 'a':", element_a)


elements = series[['b', 'c']]
print("\nElements at indices 'b' and 'c':")
print(elements)

# Accessing using position
element_0 = series.iloc[3]
print("\nElement at position 3:", element_0)

# Accessing using positions # slicing
elements_1_2_3 = series.iloc[1:4]
print("\nElements at positions 1, 2 and 3:")
print(elements_1_2_3)


```

Filter the Series to include only values greater than a specific threshold.

```python
# Sample Series
series = pd.Series([100, 50, 44, 12, 25], index=['a', 'b', 'c', 'd', 'e'])

print("Series:")
print(series)

# threshold
threshold = 30

# Filtering the Series
filtered_series = series[series > threshold]

print("\nFiltered Series (values greater than 30):")
print(filtered_series)
```

Create a DataFrame from a numpy array, specifying column and index names.

```python
# Sample array
data = np.array([
    [12, 24, 36],
    [43, 55, 46],
    [27, 58, 19]
])

print("NumPy Array:")
print(data)

# Columns
columns = ['Column1', 'Column2', 'Column3']

# Indices
index = ['Row1', 'Row2', 'Row3']

# DataFrame
df = pd.DataFrame(data, columns=columns, index=index)

print("\nDataFrame:")
print(df)
```
